```python
#Handling by droping the the row containing NaN
data=pd.read_csv('data_preprocessing.csv')
data.dropna(inplace=True)
data
```

```python
#Mean Based replacement
data=pd.read_csv('data_preprocessing.csv')
mean_Age=data['Age'].mean()
mean_Salary=data['Salary'].mean()
values={'Age':mean_Age,'Salary':mean_Salary}
data=data.fillna(value=values)
data
```

```python
#Median Based Replacement
data=pd.read_csv('data_preprocessing.csv')
median_Age=data['Age'].median()
median_Salary=data['Salary'].median()
values={'Age':median_Age,'Salary':median_Salary}
data=data.fillna(value=values)
data
```

```python
#Replacing with a Number
data=pd.read_csv('data_preprocessing.csv')
data=data.fillna(0)
data
```

```python
#Making Categories of missing data and then handling them
import pandas as pd
missing_values=[' ','--']
data=pd.read_csv('data_preprocessing.csv',na_values=missing_values)
mean_Age=data['Age'].mean()
mean_Salary=data['Salary'].mean()
values={'Age':mean_Age,'Salary':mean_Salary}
data=data.fillna(value=values)
data
```

```python
"""2.Handling categorical data"""
#One Hot Encoding
data=pd.read_csv('data_preprocessing.csv')
data=data.interpolate()
data=pd.get_dummies(data=data)
data
```

```python
#One Hot Encoding
data=pd.read_csv('data_preprocessing.csv')
data=data.interpolate()
data=pd.get_dummies(data=data,drop_first=True) #Removing the first column to get n-1 dummies out of n categorical levels 
data
```

```python
"""Splitting the data set to train data set and test data set"""
#Training data set
X_train=data.iloc[0:7,0:4]
Y_train=data.iloc[0:7,-1:]
print(X_train)
print()
print(Y_train)
```

```python
#test data set
X_test=data.iloc[7:,0:4]
Y_test=data.iloc[7:,-1:]
print(X_test)
print()
print(Y_test)
```

```python
#Standardizing the data
data=pd.read_csv('data_preprocessing.csv')
data=data.interpolate()
data=pd.get_dummies(data=data,drop_first=True) 
data['Age']=(data['Age']-data['Age'].mean())/(data['Age'].std())
data['Salary']=(data['Salary']-data['Salary'].mean())/(data['Salary'].std())
data
```

```python
"""feature Scaling"""
#Normalizing the data
data['Age']=((data['Age']-data['Age'].min())/(data['Age'].max()-data['Age'].min()))
data['Salary']=((data['Salary']-data['Salary'].min())/(data['Salary'].max()-data['Salary'].min()))
data
```
