```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn import metrics
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, SVR
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor


import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from prettytable import PrettyTable
import math


data = pd.read_csv('/home/admin1/machine_learning/week9/bike_sharing.csv')

data.drop(['instant','dteday'], axis = 1, inplace = True)


x = data.iloc[:,0:-1].values
y = data.iloc[:,-1:].values

def classification_algorithms(x, y, choice):

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
    
    if choice == 'Classification':
    
        algorithms = {LinearRegression():'LinearRegression',
                      SVC():'SupportVectorMachine', DecisionTreeClassifier():'DecisionTree', 
                      RandomForestClassifier():'RandomForest'}
    
        p = PrettyTable(["Algorithm", "Accuracy(test)", "Accuracy(train)"])

        for key,value in algorithms.items():
            
            if value == 'SupportVectorMachine':
                print(key)
    
            key.fit(x_train, y_train)
        
            y_pred_train = key.predict(x_train)
            y_pred_test = key.predict(x_test)
        
        
        
            p.add_row([value, key.score(x_test, y_test), key.score(x_train, y_train)])
        print(p)
        
        
    elif choice == 'Regression':
        
        linear_algorithms = {LinearRegression():'LinearRegression', 
                             PolynomialFeatures(degree = 2):'PolynomialRegression', 
                             SVR():'SupportVectorMachine', DecisionTreeRegressor():'DecisionTree', 
                             RandomForestRegressor():'RandomForest'}
    
        p = PrettyTable(["Algorithm","MSE", "RMSE"])

        for key, value in linear_algorithms.items():
            
            if value == 'PolynomialRegression':

                x_train_poly = key.fit_transform(x_train)
                x_test_poly = key.transform(x_test)
                
                lr = LinearRegression()
                lr.fit(x_train_poly, y_train)
                y_pred = lr.predict(x_test_poly)
                mse = metrics.mean_squared_error(y_test, y_pred)
                
                
                p.add_row([value, mse, np.sqrt(mse)])
                
                
            else:
                key.fit(x_train, y_train)
                y_pred = key.predict(x_test)
                mse = metrics.mean_squared_error(y_test, y_pred)
        
                p.add_row([value, mse, np.sqrt(mse)])
        print(p)
        
choice = input('Enter Classification for classification problems and Regression for Regression problems-')
predict = classification_algorithms(x,y, choice.capitalize())
```
