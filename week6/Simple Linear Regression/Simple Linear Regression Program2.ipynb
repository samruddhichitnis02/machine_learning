```python
"""Build a predictive linear regression model for given dataset, given humidity predict apparent temperature"""

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('weatherHistory.csv')
data.drop(['Formatted Date','Summary','Precip Type','Temperature (C)','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)','Loud Cover','Pressure (millibars)','Daily Summary'],axis=1,inplace=True)
```

```python
#Splitting the data set in to Training data set and test data set
Training_data=int(0.80*len(data))
Train_data=data.iloc[0:Training_data,:]
Test_data=data.iloc[Training_data:,:]
x_train,y_train,x_test,y_test=Train_data.iloc[:,-1:],Train_data.iloc[:,0:1],Test_data.iloc[:,-1:],Test_data.iloc[:,0:1]
```

```python
x = x_train['Humidity'].values
y = y_train['Apparent Temperature (C)'].values
```

```python
def model(x_train,y_train):
    c=0 #Theta0
    m=0 #Theta1

    L=0.01 #Learning Rate
    epoch=1000 #Number of iterations
    n=float(len(x_train))

    for i in range(epoch):
        Y_prediction=c+(m*x_train) #hypothesis function
        Derivative_c=(1/n)*sum(Y_prediction-y_train)
        Derivative_m=(1/n)*sum(x_train*(Y_prediction-y_train))
        c=c-(L*Derivative_c)
        m=m-(L*Derivative_m)
        cost=(1/(2*n))*(sum((Y_prediction-y_train)**2))
        print(cost)
    return c,m
```

```python
c,m=model(x,y)
```

```python
X = x_test['Humidity'].values
Y = y_test['Apparent Temperature (C)'].values
```

```python
y_prediction=c+(m*X)
plt.scatter(x_test['Humidity'],y_test['Apparent Temperature (C)'])
plt.plot(x_test['Humidity'],y_prediction,'r')
plt.show()
```

```python
#Calculating error
mse=np.mean((Y-y_prediction)**2)
rmse=np.sqrt(mse)
rmse
```
